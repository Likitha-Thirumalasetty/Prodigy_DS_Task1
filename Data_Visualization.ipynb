**Task 1 - Creating a histogram :** To visualize the distribution of numerical variable - here we have considered Daily Mean Travel Time

**Dataset: Uber Traffic Data Visualisation**

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
file_path ='/content/Final_Majestic_to_AIM_jan-2016tomarch-2018.xlsx'
df=pd.read_excel(file_path)
```

```python
#viewing the first five records
df.head()
```

```python
#creating a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Daily Mean Travel Time (Seconds)'], bins=20, kde=True)
plt.title('Distribution of Daily Mean Travel Time')
plt.xlabel('Daily Mean Travel Time (Seconds)')
plt.ylabel('Frequency')
plt.show()
```
